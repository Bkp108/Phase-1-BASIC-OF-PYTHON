## ✅ 11. Custom Layers & Models

```python
class MyModel(keras.Model):
    def __init__(self):
        super().__init__()
        self.dense1 = layers.Dense(64, activation='relu')
        self.dense2 = layers.Dense(10)

    def call(self, x):
        return self.dense2(self.dense1(x))
```
