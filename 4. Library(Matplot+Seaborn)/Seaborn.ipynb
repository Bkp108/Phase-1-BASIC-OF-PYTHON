## âœ… 1. Import & Setup

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()  # or sns.set_style('whitegrid')
```

---

## âœ… 2. Load Dataset

```python
df = sns.load_dataset('tips')  # other options: iris, titanic, flights, etc.
```

---

## âœ… 3. Basic Plots

### ğŸ“ˆ Line Plot
```python
sns.lineplot(x='total_bill', y='tip', data=df)
```

### ğŸ”µ Scatter Plot
```python
sns.scatterplot(x='total_bill', y='tip', data=df)
```

### ğŸŸ¨ Bar Plot (aggregated)
```python
sns.barplot(x='sex', y='tip', data=df)
```

### ğŸŸ© Count Plot (frequency)
```python
sns.countplot(x='day', data=df)
```

---

## âœ… 4. Distribution Plots

### ğŸ”º Histogram + KDE
```python
sns.histplot(df['total_bill'], kde=True)
```

### ğŸ”» KDE only
```python
sns.kdeplot(df['total_bill'])
```

### ğŸŸ¦ Box Plot
```python
sns.boxplot(x='day', y='total_bill', data=df)
```

### ğŸŸ« Violin Plot
```python
sns.violinplot(x='day', y='total_bill', data=df)
```

### ğŸŸª Strip Plot (raw data points)
```python
sns.stripplot(x='day', y='total_bill', data=df, jitter=True)
```

### ğŸŸ§ Swarm Plot (no overlaps)
```python
sns.swarmplot(x='day', y='total_bill', data=df)
```

---

## âœ… 5. Matrix Plots

### ğŸ”² Heatmap
```python
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

### ğŸ”³ Clustermap (hierarchical clustering)
```python
sns.clustermap(corr, annot=True, cmap='mako')
```

---

## âœ… 6. Categorical Relationships

```python
sns.catplot(x='day', y='total_bill', kind='box', data=df)
```

ğŸ“Œ Types: `'strip'`, `'swarm'`, `'box'`, `'violin'`, `'bar'`, `'count'`

---

## âœ… 7. Pairwise Relationships

### ğŸ” Pairplot (matrix of scatter + hist)
```python
sns.pairplot(df, hue='sex')
```

### ğŸ”‚ Jointplot
```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex')  # or 'scatter', 'kde'
```

---

## âœ… 8. Regression Plots

### ğŸ”„ Regression Line
```python
sns.regplot(x='total_bill', y='tip', data=df)
```

### ğŸ”„ Combined Facet + Regression
```python
sns.lmplot(x='total_bill', y='tip', data=df, hue='sex', col='smoker')
```

## âœ… 9. Styling
