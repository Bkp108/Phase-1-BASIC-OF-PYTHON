## ðŸ§  Bonus: Best Practices

| Practice | Why |
|---------|-----|
| Be specific in `except` | Avoid swallowing bugs |
| Use `finally` for cleanup | Always runs |
| Don't silence errors | Always log or re-raise |
| Prefer `with` statements for file/DB ops | Auto cleanup |
